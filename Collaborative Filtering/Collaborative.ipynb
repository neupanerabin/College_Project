```python
num_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
num_rating.rename(columns={"rating": "num_of_rating"},inplace = True)
```

```python
final_rating = ratings_with_books.merge(num_rating, on='title')
```

```python
final_rating = final_rating[final_rating['num_of_rating']>=50]
```

```python
final_rating.drop_duplicates(['title', 'user_id'],inplace=True)
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title',values='rating') 
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Create a KNeighborsClassifier instance
model = KNeighborsClassifier(n_neighbors=1, algorithm='brute')
```

```python
# Assuming book_sparse is your feature? matri?x and y is your target variable
# model.fit(book_sparse, y)
model.fit(book_sparse, y)
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[100, :].values.reshape(1, -1), n_neighbors=6)

```
